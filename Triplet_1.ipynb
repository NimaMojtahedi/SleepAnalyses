```python
# pattern of transition
### Pattern_3_prior--> type: all/least/top/nonactive##########
Pattern_1 = {'names':['SWS'], 'code':[-2], 'sequence':[], 'divid':3}
Pattern_2 = {'names':['REM', 'Awake'], 'code':[-3,0],'sequence':[1], 'divid':3}
Pattern_3 = {'names':['SWS', 'REM', 'SWS'], 'code':[-2, -3, -2], 'sequence':[1, 0], 'divid':3}
Pattern_3_prior = [{'name':'awake', 'code': [0], 'sequence': [1]}, {'Type' : 'nonactive', 'Percentage' : 0.2}]

# window_size (before change point and after change point)
#Window_size_ = [20,30]


# nomalization 
Nomalization_ = False # at the moment always False

# save figure address
#transition_fig_add = False

# selecting sub population
sub_pop_apply = False # or True

############################# initialize stroing variables ################################
type1_all_amp = []
type1_all_fr = []

type2_all_amp = []
type2_all_fr = []

type3_all_amp = []
type3_all_fr = []

type3_prior_all_amp = []
type3_prior_all_fr = []
###########################################################################################
############################# running code ################################################
###########################################################################################

Full_Counts = []
ro = 0
for address, sf, arts, rm_cell in zip(all_address, sfs, artifacts, remove_cell):
    # initializing ETL
    ETL_ = ETL(sf=sf, ending='.mat', unit_length=600)

    # giving file or files path
    ETL_.get_path(address)

    # loading all files to ETL memory
    ETL_.load_files()

    # getting hypno signals for all files
    Hypno_ = ETL_.get_hypno()

    # getting full data set
    full_dataset = ETL_.get_data_values()
    

    # setting artifact period
    if len(arts)!=0:
        for art in arts:
            ETL_.set_artifact(start_= art[0], end_= art[1], file_= art[2]) # start and end should be based on sampling index


    # getting artifacts
    data_artifacts, hypno_artifacts = ETL_.get_artifacts()
    
    ################ preparing data in advance ######################
    Hypnos = np.hstack(Hypno_.T)
    Data = np.hstack(full_dataset.T).T

    Hypnos_art = np.hstack(hypno_artifacts.T)
    Data_art = np.hstack(data_artifacts.T).T
    ################################################################

    if sub_pop_apply:
        sub_pop_ind = []
        sub_pop_ind = np.delete(sub_pop_index[ro], np.in1d(sub_pop_index[ro], rm_cell))

        Data = Data[:, sub_pop_ind]
        Data_art = Data_art[:, sub_pop_ind]
    else:
        Data = np.delete(Data, rm_cell, axis = 1)
        Data_art = np.delete(Data_art, rm_cell, axis = 1)
        
        
    ro += 1
    
    ##### TYPE 1 ########
    
    temp_amp = []; temp_fr = []
    Trip1 = Triplet(data = Data, hypno = Hypnos, data_artifact=Data_art, hypno_artifact=Hypnos_art , sf = sf, 
            pattern=Pattern_1, normalize=Nomalization_)
    # claculating triplet 
    temp_amp, temp_fr = Trip1.type1()
    type1_all_amp.append(temp_amp)
    type1_all_fr.append(temp_fr)
    
    
    ##### TYPE 2 #######
    temp_amp = []; temp_fr = []
    Trip2 = Triplet(data = Data, hypno = Hypnos, data_artifact=Data_art, hypno_artifact=Hypnos_art , sf = sf, 
            pattern=Pattern_2, normalize=Nomalization_)
    # claculating triplet 
    temp_amp, temp_fr = Trip2.type2()
    type2_all_amp.append(temp_amp)
    type2_all_fr.append(temp_fr)
    
    
    ##### TYPE 3 #######
    temp_amp = []; temp_fr = []
    Trip3 = Triplet(data = Data, hypno = Hypnos, data_artifact=Data_art, hypno_artifact=Hypnos_art , sf = sf, 
            pattern=Pattern_3, normalize=Nomalization_)
    # claculating triplet 
    temp_amp, temp_fr = Trip3.type3()
    type3_all_amp.append(temp_amp)
    type3_all_fr.append(temp_fr)
    
    
    ##### TYPE 3 Prior #######
    temp_amp = []; temp_fr = []
    Trip3 = Triplet(data = Data, hypno = Hypnos, data_artifact=Data_art, hypno_artifact=Hypnos_art , sf = sf, 
            pattern=Pattern_3, normalize=Nomalization_)
    # claculating triplet 
    temp_amp, temp_fr = Trip3.type3_prior_class(prior_class=Pattern_3_prior[0], 
                                                subset=Pattern_3_prior[1])
    type3_prior_all_amp.append(temp_amp)
    type3_prior_all_fr.append(temp_fr)
    
    print('*'*25 , '    MOVING TO NEXT   ', '*'*25)

```

```python
def create_stats_data(triplet_data, pattern_names, nr_divid, mouse_nr = 1):
    cols = []
    for i, pattern in enumerate(pattern_names):
        for ii in range(nr_divid):
            cols.append(pattern + '_' + str(ii+1) +'th')

    df = pd.DataFrame(columns=cols.append('mouse_nr'))
    for i in range(len(pattern_names) * nr_divid):
        df[cols[i]] = triplet_data[:,i]
        
    df['mouse_nr'] = mouse_nr
    return df
```
